# NYC 311 Service Requests

Chapter 2

```python
complaints = pd.read_csv('../data/311-service-requests.csv', dtype='unicode')
complaint_count = complaints['Complaint Type'].value_counts()
complaint_count[:10].plot(kind='bar')
```

Chapter 3

```python
# Noise - Street/Sidewalk complaints
noise_complaints = complaints[complaints['Complaint Type'] == 'Noise - Street/Sidewalk']
# Noise - Street/Sidewalk complaints grouped by borough 
noise_complaints['Borough'].value_counts().plot(kind='bar')
# Noise - Street/Sidewalk complaints grouped by borough as percent of complaints in borough
noise_complaint_counts = noise_complaints['Borough'].value_counts()
complaint_counts = complaints['Borough'].value_counts()
(noise_complaint_counts / complaint_counts).plot(kind='bar')
```
